```python
import pandas as pd
```

```python
gymdf = pd.read_csv('2022 Exercise establishment data from CBP.csv')
#downloaded from https://data.census.gov/table?q=CBP2022.CB2200CBP&g=010XX00US$0500000_050XX00US01001&y=2022&n=621340:711211:711219:712190:713920:713930:713940:713950:713990:813410

gymdf['fips_code'] = gymdf['GEO_ID'].str[-5:]


```

```python
gymdf = gymdf[['fips_code', 'ESTAB', 'PAYANN', 'EMP']]
gymdf.columns
```

```python
gymdf["ESTAB"] = pd.to_numeric(gymdf["ESTAB"], errors='coerce')
gymdf["PAYANN"] = pd.to_numeric(gymdf["PAYANN"], errors='coerce')
gymdf["EMP"] = pd.to_numeric(gymdf["EMP"], errors='coerce')


df_grouped = gymdf.groupby("fips_code").agg({
    "ESTAB": "sum",  # Total establishments
    "PAYANN": "sum",  # Total annual payroll
    "EMP": "sum"      # Total employees
}).reset_index()
df_grouped.head(10)
```

```python
chr = pd.read_csv('CHR2023.csv')
print(chr.columns.tolist())
chr['fips_code'] = chr['5-digit FIPS Code'].astype(str).str.zfill(5)
```

```python
columns_to_keep = [
    # Core Geography
    'fips_code', 'State Abbreviation', 'Name', 'Population raw value',
    
    # Diabetes & Direct Health Factors
    'Diabetes Prevalence raw value', 'Adult Obesity raw value', 
    'Physical Inactivity raw value', 'Food Environment Index raw value',
    'Poor or Fair Health raw value', 'Limited Access to Healthy Foods raw value',
    
    # Economic & Inequality
    'Median Household Income raw value', 'Income Inequality raw value',
    'Children in Poverty raw value', 'Unemployment raw value',
    'Living Wage raw value', 'Child Care Cost Burden raw value',
    
    # Education & Opportunity
    'High School Graduation raw value', 'Some College raw value',
    'Disconnected Youth raw value', 'Reading Scores raw value',
    
    # Environmental
    'Air Pollution - Particulate Matter raw value', 
    'Drinking Water Violations raw value',
    'Severe Housing Problems raw value', 'Broadband Access raw value',
    'Traffic Volume raw value',
    
    # Behavioral/Social
    'Excessive Drinking raw value', 'Adult Smoking raw value',
    'Social Associations raw value', 'Voter Turnout raw value',
    'Homeownership raw value',
    
    # Demographics
    '% Rural raw value', '% 65 and Older raw value'
]
```

```python
chr = chr[columns_to_keep]
chr.columns = [col.replace(' raw value', '') for col in chr.columns]
chr.columns = [col.replace(' ', '_') for col in chr.columns]
chr.head(10)
```

```python
chr_and_gym = pd.merge(
    chr, 
    gymdf, 
    on='fips_code',          # Common column
    how='inner',             # Merge type
    suffixes=('_gym', '_chr')  # Suffixes if columns overlap
)
chr_and_gym.head(10)
```

```python
chr_and_gym.columns
```

```python
svi = pd.read_csv('svi_interactive_map.csv', dtype={'FIPS': str})
#downloaded from https://www.atsdr.cdc.gov/place-health/php/svi/svi-interactive-map.html
svi.head()
```

```python
svi.rename(columns={'FIPS': 'fips_code', 'RPL_THEMES': 'Percentile_SVI'}, inplace=True)
svi.head()
```

```python
svi=svi[['fips_code', 'Percentile_SVI']]
svi.head()
```

```python
chr_gym_SVI = pd.merge(
    chr_and_gym, 
    svi, 
    on='fips_code',
)
chr_gym_SVI.head(10)
```

```python
chr_gym_SVI.to_csv('chr_gym_SVI.csv', index=False)
```
